```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#heatmap for all constructs pulsing
sns.set()
data = pd.read_csv("heatmapparameters_FBLC_smaller_50.csv")

df = pd.DataFrame(data, columns=["d2"])

y_axis_labels = ['Ies1p','Hir3p','Btt1p','Caf4p','Bur2p','Rad6p','CIB1','Rtt102p','Chz1p','Nut1p','Sus1p','Cpr1p','Med8p','Med6p','Hpa2p','VP16']
x_axis_labels = ['d2 [s$^{-1}$]']
sns.set(font_scale=2)
plt.subplots(figsize=(10,10))
heatmap = sns.heatmap(df,xticklabels=x_axis_labels,yticklabels=y_axis_labels, linewidth=0.05,vmax=40, vmin=5, cmap="viridis",square=True)

figure = heatmap.figure
plt.ylabel('Time (min)',labelpad=15)
plt.show()

figure.savefig("d2heatmap_FBLC_smaller_50_reordered.svg",bbox_inches='tight')


#heatmap for all constructs pulsing
fig, ax = plt.subplots(figsize=(10,10))

sns.set()
data = pd.read_csv("heatmapparameters_FBLC_smaller_50.csv")

df = pd.DataFrame(data, columns=["k7"])

y_axis_labels = ['Ies1p','Hir3p','Btt1p','Caf4p','Bur2p','Rad6p','CIB1','Rtt102p','Chz1p','Nut1p','Sus1p','Cpr1p','Med8p','Med6p','Hpa2p','VP16']
x_axis_labels = ['k7 [s$^{-1}$]']
sns.set(font_scale=2)
heatmap = sns.heatmap(df,xticklabels=x_axis_labels,yticklabels=[], linewidth=0.05,vmax=40, vmin=5, cmap="viridis",square=True)

figure = heatmap.figure
#plt.ylabel('Time (min)',labelpad=15)
plt.show()

figure.savefig("k7heatmap_FBLC_smaller_50_reordered.svg",bbox_inches='tight')
```
